```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Takes input from the file and creates a graph using networkx library
def CreateGraph():
    G = nx.Graph()
    f = open('newgraph.txt')
    for line in f:
        if line.startswith("e "):
            graph_edge_list = line.split()
            G.add_edge(graph_edge_list[1], graph_edge_list[2]) 
    return G
```

```python
#draws the graph and displays the weights on the edges
def DrawGraph(G,col_val):
    pos = nx.spring_layout(G)
    values = [col_val.get(node, 'blue') for node in G.nodes()]
    nx.draw(G, pos, with_labels = True, node_color = values, edge_color = 'black' ,width = 0.5, alpha = 0.4, 
            cmap="viridis", label="Graph Coloring", node_size = 1000, font_size = 20,
           font_weight="bold")  #with_labels=true is to show the node number in the output graph
```

```python
G = CreateGraph()
```

```python
d = {}
with open("solution.txt") as f:
    for i in range(1):
        line = f.readline()
    for line in f:
        (key, val) = line.split()
        d[key] = val
```

```python
d
```

```python
DrawGraph(G,d)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.title('Graph Coloring of graph.txt', fontsize=30)
```
